````python
import openai

# Set your API key
client = OpenAI(api_key="Your OpenAPI key")

# Function to analyze sentiment using OpenAI
def create_embeddings(texts):
    """This is a function to create embedding(s) by OpenAI API

    Args:
        texts (string): the string will be dealt with

    Returns:
        list: the list of embedding(s)
    """
    response = client.embeddings.create(
        model="text-embedding-ada-002",
        input=texts
    )
    response_dict = response.model_dump()
    
    return [data['embedding'] for data in response_dict['data']]
````

````python
# Only print first n and last n data
def print_list_of_embeddings(embeddings, n=2, level=0):
    """This is a function used to print the first and last n numbers of embeddings
    list generated by OpenAI API which has 1536 float values.

    Args:
        embeddings (_type_): the list generated by OpenAI API which has 1536 float values 
        n (int, optional): printed number of the values. Defaults to 2.
        level (int, optional): to adjust the indentation. Defaults to 0.
    """
    indentation = "  "*level
    print(indentation + "embeddings: [")
    if len(embeddings) > 2 * n:
        for i in range(n):
            print(indentation + "  ", embeddings[i], ",")
        print(indentation + "    ...")
        for i in range(len(embeddings) - n, len(embeddings)):
            print(indentation + "  ", embeddings[i], ",")
    else:
        for i in range(len(embeddings)):
            print(indentation + "  ", embeddings[i], ",")
    print(indentation + " ]")

def print_dict_with_partial_embeddings(response_dict, n=2):
    """This is a function used to print the first and last n numbers of embeddings
    list in a dictionary generated by OpenAI API

    Args:
        response_dict (dict): the dictionary generated by client.embeddings.create.model_dump()
        n (int, optional): printed numbers of the embeddings. Defaults to 2.
    """
    for key, value in response_dict.items():
        if key == 'data' and isinstance(value, list):
            print(f"{key}: [")
            for item in value:
                print(" {")
                for sub_key, sub_value in item.items():
                    if 'embedding' in sub_key and isinstance(sub_value, list):
                        print_list_of_embeddings(item['embedding'], n, 1)
                    else:
                        print(f"  {sub_key}: {sub_value},")
                print(" },")
            print("]")
        elif key == 'usage' and isinstance(value, dict):
            print(f"{key}: {{")
            for sub_key, sub_value in value.items():
                print(f"  {sub_key}: {sub_value}")
            print("}")
        elif key == 'embedding' and isinstance(value, list):  
            print_list_of_embeddings(value, n)  
        else:
            print(f"{key}: {value}")
````
